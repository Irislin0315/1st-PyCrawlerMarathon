### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
#print(soup.find(class_="part_list_2").find_all('h3')[0])
#print(soup.find(class_="part_list_2").find_all('h3')[1].find('em').text)
for i in soup.find(class_="part_list_2").find_all('h3'):
    print(i.find(class_="date").text, i.find('em').text, i.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
date=[]
category=[]
title=[]
for i in soup.find(class_="part_list_2").find_all('h3'):
    date.append(i.find(class_="date").text) 
    category.append(i.find('em').text)
    title.append(i.find_all('a')[-1].text)
dict_news = {'Date':date,'Category':category,'Title':title}
df_news = pd.DataFrame(dict_news)
df_group = df_news.groupby(df_news['Category'])
```

### ③ 哪一個類別的文章最多

```python
print(df_group.size())
```
